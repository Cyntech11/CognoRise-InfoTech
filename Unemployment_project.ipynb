**EXPLORATORY DATA ANALYSIS**

**The rows has been reduced from 768 to 740**

```python
df.describe()
```

```python
from datetime import datetime
df[' Date'] = pd.to_datetime(df[' Date'])

df.info()
```

```python
df.rename(columns={' Date':'Date', ' Frequency':'Frequency', ' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)', ' Estimated Employed':'Estimated Employed', ' Estimated Labour Participation Rate (%)':'Estimated Labour Participation Rate (%)'},inplace=True)
```

```python
df.select_dtypes(include='object').columns
```

```python
df.select_dtypes(include='number').columns
```

```python
df['Estimated Employed'] = round(df['Estimated Employed']/1000000,2)

df.head()
```

```python
df.duplicated().sum()
```

```python
df.Frequency.value_counts()
```

**So let's also note that the Frequency column has no relevant because we can extract the months from our date column......so let's drop it**

```python
df.drop('Frequency',axis=1,inplace=True)
```

```python
df.head()
```

**we want to do some Feature engineering.......So let's extract the month and year from the date column....**

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

df.head()
```

**Analyzing the Area Column**

```python
df['Area'].value_counts()
```

```python
px.pie(df,names='Area',title='Count Plot of Area')
```

**Let's note that the Unemployment rate for Urban is slightly higher than the rural area**

```python
Region = df.groupby(['Region'])['Estimated Unemployment Rate (%)'].mean().reset_index()

Region

px.bar(Region,x='Region',y='Estimated Unemployment Rate (%)',title='Average Unemployment Rate by Region')
```

```python
px.bar(df,x='Region',y='Estimated Unemployment Rate (%)',title='Average Unemployment Rate by Region')
```
